```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv("integración.csv")
data.drop("#",axis=1,inplace=True)
data.head(15)
```

```python
data["Escuela"].head()
```

```python
dups = data.pivot_table(index = ['Escuela'], aggfunc ='size')
dups
```

```python
dups.index.values[0]
```

```python
seleccion=dups.index.values[0]
seleccion
```

```python
escuela=data[(data.Escuela==seleccion)]
escuela.head(20)
```

```python
escuela.groupby("SEXO").count()
```

```python
escuela.groupby("indice de Masa Corporal").count()
```

```python
escuela.groupby("SEXO").count()["indice de Masa Corporal"].plot(kind="bar")
```

```python
escuela.groupby("indice de Masa Corporal").count()["SEXO"].plot(kind="bar")
```

```python
por_masa=pd.crosstab(escuela["indice de Masa Corporal"],escuela["SEXO"])
por_masa
```

```python
por_masa.columns
```

```python
por_masa.index
```

```python
graf_masa=por_masa.plot(kind="bar",figsize=(8,6),title="Índice de masa corporal por sexo")
```

```python
fig = graf_masa.get_figure()
fig.savefig("output.png")
```
